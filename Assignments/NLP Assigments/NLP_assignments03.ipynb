### 1. Explain the basic architecture of RNN cell.

In simple feedforward neural networks, where the activations flow only in one direction, from the input layer to the output layer.

Where the a recurrent neural network looks very much like a feedforward neural network, except it also has connections point backward. 

In this cell, one neuron receiving inputs, producig an output, and sending that output back to itself.


At each **time step t**(also called a frame). This recurrent neuron receives the inputs **x(t)** as weel as its own output from the previous time step, **y(t-1)**.

**Both the inputs and outputs are vectors**

Each recurrent neuron has two sets of weights: one for the inputs *x(t)* and the other for the outputs of the previous time step *y(t-1)*.


### 2. Explain Backpropagation through time (BPTT)

To train an RNN, the trick is to unroll it through time and then simply use regular backpropagation. This strategy is called *backpropagation through time*



In regular backpropagation, There is a first forward pass through the unrolled network(dashed arrows). Then the output sequence is evaluated using a cost function(C(Y0, Y1, ....YT))

### 3. Explain Vanishing and exploding gradients

The backpropagation algorithm(update weights) works by going from the output layer to the input layer. 

Once the algorithm has computed the gradient of the cost function with regard to each parameter in the network, it uses these gradients to update each parameter with Gradient Descent step.

While the gradients gets smaller and smaller as the algorithm progresses down to the lower layers. These weights gets smaller because when we backpropagate we will be derivating the weights, in generally sigmoid activation function(ranges between 0 to 1). When derivative these function may ranges between 0 to 0.25. That's why it gets smaller. This is called **Vanishing Gradients** problem.

In some cases, the opposite can happen the gradients can grow bigger and bigger until layers get insanely large weight updates the algorithm diverges. This is called **exploding gradients** problem.

### 4. Explain Long short-term memory (LSTM)

In RNN, the major problem is *short-term memory* problem. Where it leaves(forgets) some important context while training. 


While backpropagation it also gets vanishing gradient problem.

To overcome the above issue we use **LSTM(Long-Short Term Memory)**.
LSTM are explicitly designed to avoid the long-term dependency problem. Remembering information for long periods of time is practically their default behaviour, not something they struggle to learn.


LSTMs also have the chain like structure, but the repeating module has a different structure.

### 5. Explain Gated recurrent unit (GRU)

The GRU(Gated Recurrent Unit) cell is a simplified version of the LSTM cell.
GRU has two gates
1. Reset gate(short-term memory)
2. Update gate(long-term memory)


### 6. Explain Peephole LSTM

In a regular LSTM cell, the gate controllers can look only at the input(Xt) and previous short-term state(h(t-1)). ### 9. Explain BiLSTM

A  Bidirectional LSTM, or biLSTM, is a sequence processing model that consists of two LSTMs: one taking the input in a forward direction, and the other in a backwards direction. BiLSTMs effectively increase the amount of information available to the network, improving the context available to the algorithm (e.g. knowing what words immediately follow and precede a word in a sentence).
