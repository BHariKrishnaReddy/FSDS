```python
import seaborn as sns

# Load the "tips" dataset from seaborn
tips = sns.load_dataset("tips")

# Create a boxen plot of tip amount by day and time
g = sns.catplot(x="day", y="tip", hue="time", kind="boxen", data=tips)

# Set the title and axis labels
g.fig.suptitle("Tip Amount by Day and Time")
g.set_xlabels("Day of the Week")
g.set_ylabels("Tip Amount")
```

```python
import pandas as pd
import seaborn as sns

# Create a sample dataframe
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2010, 2011, 2012, 2010, 2011, 2012],
    'Month': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'Value': [10, 20, 30, 15, 25, 35, 12, 22, 32]
})

# Group the data by year and month and calculate the mean value
grouped_data = data.groupby(['Year', 'Month']).mean()

# Convert the grouped data to a pivot table
pivot_table = grouped_data.pivot_table(values='Value', index='Year', columns='Month')

# Plot the heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
```
