### Changing everything into machine readable and filling null values

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
df['Age']=df['Age'].map({'0-17':0, '55+':1, '26-35':6,'46-50':3, '51-55':2, '36-45':5, '18-25':4})
df['City_Category']=df['City_Category'].map({'A':0,'B':1,'C':2})
df['Stay_In_Current_City_Years']= df['Stay_In_Current_City_Years'].map({'0':0,'1':1,'2':2,'3':3,'+4':4})
```

Findging Eigenvectors and EigenValues
for this we need to remove null values from Product_Category_2 and Product_Category_3

#### Handling with Median

```python
df["Stay_In_Current_City_Years"].fillna(df["Stay_In_Current_City_Years"].median(), inplace=True)
df["Product_Category_2"].fillna(df["Product_Category_2"].median(), inplace=True)
df["Product_Category_3"].fillna(df["Product_Category_3"].median(), inplace=True)
```

```python
df3= np.array(df)
covarMat = np.cov(df3.T) #np.cov() can only deal with array data types
eigenValues, eigenVectors = np.linalg.eig(covarMat)
print('Values:',eigenValues)
print('\n Vectors',eigenVectors)
```

```python
for ev in eigenVectors.T:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
    
# Make a list of (eigenvalue, eigenvector) tuples
eigPairs = [(np.abs(eigenValues[i]), eigenVectors[:,i]) for i in range(len(eigenValues))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigPairs.sort(key=lambda x: x[0], reverse=True)
for i in eigPairs:
    print(i[0])
```

```python
pca = PCA(n_components=10)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents , columns = ['pcaf1', 'pcaf2', 'pcaf3', 'pcaf4','pcaf5', 'pcaf6', 'pcaf7', 'pcaf8','pcaf9', 'pcaf10'])
```

```python
# Perform PCA and select the top k eigenvectors
pca = PCA(n_components=6)
X_pca = pca.fit_transform(df.drop('Purchase',axis=1))

# Print the explained variance ratio for each component
print(pca.explained_variance_ratio_)

# Print the total explained variance
print(sum(pca.explained_variance_ratio_))
```
